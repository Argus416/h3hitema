# Visualisation d’un nuage de points et exploration des variables

Pour cette partie, nous allons utiliser le dataset suivante : house_pricing.csv que vous pouvez trouver dans le dossier Data Sources. Nous allons tenter de comprendre un peu mieux ce dataset par la visualisation 

## Importez les librairies qu’il nous faut (pandas, seaborn, matplotlib, numpy)

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib as mpot
```

## Donnez la liste des variables présentes dans ce dataset ainsi que leur nature (sont elles qualitatives, quantitatives, discrète etc...) et leur type (float, int, str etc...)


```python
csvFilePath = "./Data/house_pricing.csv"

df = pd.read_csv(csvFilePath)

df.info()
```

## En créant un nuage de points, regardez comment se comporte la colonne LotArea par rapport au SalesPrices

```python
sns.set_theme()

sns.relplot(
    data=df,
    x="LotArea", y="SalePrice", marker="x", 
)


```

## Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea inférieur à 20 000 pieds carrés et un prix inférieur à 500 000$ 

```python
df = df[ (df.LotArea < 20000) & (df.SalePrice < 500000)]; df
```

## En créant un nuage de points, regardez la relation entre le LotFrontage et le LotArea

```python
sns.relplot(
    data=df,
    x="LotArea", y="SalePrice"
)
```

## De la même manière, affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotFrontage inférieur à 200 pieds carrés et un LotArea inférieur à 100000 pieds carré

```python
df = df[(df.LotFrontage < 200) & (df.LotArea < 100000)]
```

```python
sns.relplot(
    data=df,
    x="LotArea", y="SalePrice", s = 20, height= 10
)
```

```python
df = df.iloc[:,-16:-1]

df
```

```python
matriceDeCorrelation = df.corr(method='pearson')
```

```python
sns.heatmap(data=matriceDeCorrelation, annot=True)
```
